```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
```

```python
df = pd.read_excel("finch_tunnel_data.xlsx", sheet_name="Combined")
BW_df = pd.read_excel("finch_tunnel_data.xlsx", sheet_name="Black-White")
WB_df = pd.read_excel("finch_tunnel_data.xlsx", sheet_name="White-Black")

BW_df = BW_df[BW_df.Entry.isna() != True]
WB_df = WB_df[WB_df.Entry.isna() != True]

```

```python
# ENTRY POINT HISTOGRAMS
BW_df.hist(column='Entry')
WB_df.hist(column='Entry')
```

```python
# TRAJECTORY BAR PLOTS
#sns.countplot(x="Origin", hue="Direction", data=df)
sns.countplot(x="Treatment", hue="Direction", data=df)
plt.show()
sns.countplot(x="Treatment", hue="Origin", data=df)
plt.show()
sns.countplot(x="Origin", hue="Direction", data=df)
plt.show()
```

```python
# ORIGIN CAT PLOTS
sns.catplot(x="Origin", y="Entry", hue="Treatment", kind="box", data=df)
```

```python
# DIRECTION CAT PLOTS
sns.catplot(x="Direction", y="Entry", hue="Treatment", kind="box", data=df)
```
